```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/mcs/Downloads/spam.csv', encoding='latin-1')
```

```python
df = df[['v1', 'v2']]
```

```python
df.columns = ['label', 'msg']
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
nltk.download('punkt')
```

```python
def extract_features(msg):
    capital_words = re.findall(r'\b[A-Z]+\b', msg)

    numerals = re.findall(r'\b\d+\b', msg)

    special_chars = re.findall(r'[^\w\s]', msg)

    words = word_tokenize(msg)

    one_letter_words = [word for word in words if len(word) == 1 and word.isalpha()]

    multi_letter_words = [word for word in words if len(word) > 1]

    sentences = sent_tokenize(msg)

    if sentences:
        avg_sentence_length = sum(len(word_tokenize(sent)) for sent in sentences) / len(sentences)
    else:
        avg_sentence_length = 0

    return {
        'capital_words': capital_words,
        'numerals': numerals,
        'special_chars': special_chars,
        'one_letter_words': one_letter_words,
        'multi_letter_words': multi_letter_words,
        'avg_sentence_length': avg_sentence_length
    }

df['features'] = df['msg'].apply(extract_features)
df_features = pd.json_normalize(df['features'])
df_combined = pd.concat([df, df_features], axis=1)

# print(df_combined[['msg', 'capital_words', 'numerals', 'special_chars', 'one_letter_words', 'multi_letter_words', 'avg_sentence_length']].head())

for index, row in df_combined[['msg', 'capital_words', 'numerals', 'special_chars', 'one_letter_words', 'multi_letter_words', 'avg_sentence_length']].iterrows():
    print(f"msg: {row['msg']}\n"
          f"capital_words: {row['capital_words']}\n"
          f"numerals: {row['numerals']}\n"
          f"special_chars: {row['special_chars']}\n"
          f"one_letter_words: {row['one_letter_words']}\n"
          f"multi_letter_words: {row['multi_letter_words']}\n"
          f"avg_sentence_length: {row['avg_sentence_length']}\n")
```
